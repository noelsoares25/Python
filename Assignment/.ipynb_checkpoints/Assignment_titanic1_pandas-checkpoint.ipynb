```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'C:\Users\NOEL\Desktop\Data Science\Class\titanic_dataset.csv')
data=pd.DataFrame(df)
df.describe()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
print(df.isna().sum()*100/df.shape[0]>25)
```

```python
data.drop(columns=['cabin','boat','body','home.dest'],inplace=True)
```

```python
data.drop(columns=['ticket','name'],inplace=True)
```

```python
data.dtypes
```

```python
data['fare']=data['fare'].str.replace('$', '')
```

```python
data
```

```python
data['fare']=pd.to_numeric(data.fare)
```

```python
data.dtypes
```

```python
data.head()
```

```python
print(data.isna().any())
print()
print(data.isna().sum())
```

```python
data.fare.fillna(data.fare.mean(),inplace=True)
data.age.fillna(data.age.mean(),inplace=True)
print(data.isna().sum())
```

```python
data.dropna(subset=['embarked'],inplace=True)
print(data.isna().sum())
```

```python
data.to_csv('titanic_filtered.csv')
```

```python
data.survived.replace({0:'D',1:'A'},inplace=True)
print(data.survived.head())
```

```python
data.survived.value_counts()
```

```python
g1=data.groupby(['gender','survived'])
g1.count()
```

```python
print(data.pclass.unique())
print()
print(data.pclass.value_counts())
```

```python
a=data.nlargest(5,'age')
print(a)
print(a.gender.value_counts())
```

```python
sur=data[(data.survived=='A') & (data.gender=='male')]
print(sur.nlargest(1,'age'))
print()
sur=data[(data.survived=='A') & (data.gender=='female')]
print(sur.nlargest(1,'age'))
```

```python
avg=data.groupby('gender').age.mean()
print(avg)
```

```python
avg=data.groupby('survived').age.mean()
print(avg)
```
